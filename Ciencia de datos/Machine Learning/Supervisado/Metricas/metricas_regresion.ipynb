# Mtricas de regresin

Una regresin es una funcin que nos permite predecir un valor en funcin de otras variables, para definir que tan buenas son estas predicciones, se calculan ciertas mtricas.



# Error absoluto mximo

$y:$ Valor verdadero

$\hat y:$ Valor predicho

$$M(y,\hat y)=max(|y_i-\hat y_i|)$$

```python
from sklearn.metrics import max_error

y = [1, 2, 3, 4, 5]
y_pred = [1, 2, 3, 4, -5]
max_error(y, y_pred)
```

# Error absoluto medio MAE

$y:$ valor verdadero

$\hat y:$ valor predicho

$n:$ tamao de muestra

$$MAE(y,\hat y)=\frac{1}{n}\sum_{i=1}^n|y_i-\hat y_i|$$

```python
from sklearn.metrics import mean_absolute_error

y = [1, 2, 3, 4, 5]
y_pred = [1, 2, 3, 4, -5]
mean_absolute_error(y, y_pred)
```

# Error cuadratico medio MSE

$y:$ valor verdadero

$\hat y:$ valor predicho

$n:$ tamao de muestra

$$MSE(y,\hat y)=\frac{1}{n}\sum_{i=1}^n(y_i-\hat y_i)^2$$

Castiga ms los fallos grandes

```python
from sklearn.metrics import mean_squared_error

y = [1, 2, 3, 4, 5]
y_pred = [1, 2, 3, 4, -5]
mean_squared_error(y, y_pred) * len(y)
```

# Suma de los cuadrado de los resudios RSS

$y:$ valor verdadero

$\hat y:$ valor predicho

$n:$ tamao de muestra

$$RSS(y,\hat y)=\sum_{i=1}^n(y_i-\hat y_i)^2$$

```python
from sklearn.metrics import mean_squared_error

y = [1, 2, 3, 4, 5]
y_pred = [1, 2, 3, 4, -5]
mean_squared_error(y, y_pred)
```

# Raz cuadrada del error cuadratico medio RMSE

$y:$ valor verdadero

$\hat y:$ valor predicho

$n:$ tamao de muestra

$$MSE(y,\hat y)=\frac{1}{n}\sum_{i=1}^n(y_i-\hat y_i)^2$$

```python
from sklearn.metrics import root_mean_squared_error

y = [1, 2, 3, 4, 5]
y_pred = [1, 2, 3, 4, -5]
root_mean_squared_error(y, y_pred)
```

# Coeficiente de determinacin R2

$y:$ valor verdadero

$\hat y:$ valor predicho

$\bar y:$ promedio de los valores verdaderos

$n:$ tamao de muestra

$$R^2(y,\hat y)=1-\frac{\sum_{i=1}^n(y_i-\hat y_i)^2}{\sum_{i=1}^n(y_i-\bar y)^2}$$

No aplica para modelos de regresin NO lineales

```python
from sklearn.metrics import r2_score

y = [1, 2, 3, 4, 5]
y_pred = [1, 2, 3, 4, 3]
r2_score(y, y_pred)
```

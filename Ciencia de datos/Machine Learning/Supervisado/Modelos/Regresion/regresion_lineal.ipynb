# Regresin lineal

Es uno de los modelos ms senciloos para predecir una variable en funcin de otra. Este modelo es extrado de estadstica, su frmula es:

$$\hat y = B_0+B_1 X_1+B_2 X_2+\dots+B_n X_n$$



## Supocisiones

- La variable dependiente est linealmente relacionada con las variables independientes.
- Las variables independientes no estn correlacionadas entre s.
- Las variables independientes no estn correlacionadas con el error del modelo.
- Los errores del modelo son independientes entre s.
- Los errores del modelo tienen varianza constante.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../../../../../data/raw/ingreso.csv")
data.head()
```

```python
sns.regplot(data, x="horas", y="ingreso")
```

## Creacin del modelo

```python
X_train, X_test, y_train, y_test = train_test_split(data[["horas"]], data["ingreso"], test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f"Su formula es: {model.intercept_:.2f} + {model.coef_[0]:.2f}")
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
plt.plot(X_test, y_pred, "r", label="Predicciones")
plt.scatter(X_test, y_test, label="Reales")
plt.legend()
plt.show()
```

```python
r2_score(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
